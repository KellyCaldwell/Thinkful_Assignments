```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('bikepghpublic.csv')
df
```

```python
plt.plot(df['SafetyHuman'], color='red')
plt.ylim([-0.1, 5.0])
plt.ylabel('Perceived Human Safety')
plt.xlabel('Participant By Number')
plt.title('Pittsburgh Public AV Survey')
plt.show()
```

```python
plt.scatter(x=df['FamiliarityTechnoology'], y=df['ZipCode'])
plt.ylim([15125, 15275])
plt.ylabel('ZipCode')
plt.xlabel('Participant\'s AV Familiarity')
plt.title('Pittsburgh Public AV Survey')
plt.show()
```

```python
plt.plot(df['AVSafetyPotential'], color='blue')
plt.ylabel('Perceived AV Safety Potential')
plt.xlabel('Participant By Number')
plt.title('Pittsburgh Public AV Survey')
plt.show()
```

```python
plt.plot(df['FamiliarityTechnoology'], color='green')
plt.ylabel('Familiarity With Technology')
plt.xlabel('Participant By Number')
plt.title('Pittsburgh Public AV Survey')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['AVSafetyPotential'], color='blue')
plt.ylabel('AV Safety Potential')
plt.title('AV Safety Potential')

plt.subplot(1, 2, 2)
plt.plot(df['FamiliarityTechnoology'], color='green')
plt.ylabel('Familiarity With Technology')
plt.title('Familiarity With Technology')
plt.show()
```
